```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os
import glob


headers = {
    'User-Agent': 'Mozilla/5.0 (Linux; Android 5.1.1; SM-G928X Build/LMY47X) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.83 Mobile Safari/537.36'}
```

```python
url = 'https://www.dol.gov/agencies/eta/foreign-labor/performance'
baseurl = 'https://www.dol.gov'
page = requests.get(url=url, headers=headers)
#print(page.text)
soup = BeautifulSoup(page.content,"html.parser")
#print(soup.prettify())

fiscal_year = ['FY'+ str(x) for x in range(2017, 2020)] + ['FY'+ str(x)+'_Q' + str(y) for x in range(2020, 2022) for y in range(1,5)]

files = []
table = soup.find("table", attrs = {'class' : 'cust_table','summary' : 'LCA Programs (H-1B, H-1B1, E-3)'})
table_body = table.find('tbody')               
for row in table.findAll("tr"):
    for link in row.findAll('a'):
        if link.text.endswith('xlsx'):
            file_name = (re.findall('FY\d{4}[\_Q\d{1}]*', link.text))[0]
            if  file_name in fiscal_year:
                files.append([file_name ,baseurl + link['href']])
files[0]
```

```python
print("Downloading...")
for file in files:
    resp = requests.get(file[1])
    filename = file[0]
    with open(os.path.join(os.getcwd(),'Data/', f'{filename}.csv'), 'wb') as output:
        output.write(resp.content)
print("Done!")
```
